Para selecionar linhas especificas fazemos uso do loc[] o primeiro valor determina a linha escolhida [1], [1:3] = vai da linha de indice 1 até 3, [:3] = vai do inicio do dataframe até a linha 3, [3:] a patir da 3 até a ultima

o loc também seleciona colunas especificas loc[[:3], 'Produto', 'Quantidade']]

```python
display(planilha_vendas.loc[:3])
```

# A adição de colunas
Deve ser feita após unir planilhas, do contrário aparecera valores NaN que são celulas sem valores

Excluindo colunos e linhas
colunas = drop['nome da coluna', axis = 1], o valor 1 do axis informa ao panda que é uma coluna
linhas = drop[1, axis = 0], o valor 0 do axis informa ao panda que é uma linha

```python
venda_ano = venda_ano.drop('Imposto', axis = 1)
```

```python
display(venda_ano)
```

```python
venda_ano = venda_ano.drop(3, axis = 0)
```

```python
display(venda_ano)
```

# Valores vazios
- Deletar linhas que estão completamente vazias
    - venda_ano = venda_ano.drop(how 'all' , axis = 0), por padrão se não adicionar o axis = 0 o pandas já iria deletar as linhas
- Deletar colunas que estçao completamente vazias
    - venda_ano = venda_ano.drop(how 'all', axis = 1), axis = 1 indica coluna
- Deletar linhas que possuem pelo menos um valor vazio
    - venda_ano = venda_ano.dropna()
- Preencher valores vazios com a media ou ultimo valor
    - media dos valores da coluna, venda_ano['Comissão'] = venda_ano['Comissão'].fillna(venda_ano['Comissão'].mean()).
    no .fillna() especificamos o valor a ser atribuido nos espaços vazios, se atribuir .fillna(1) os espaços vazios seram preenchidos com o valor 1.
    - valor antecessor venda_ano = venda_ano.ffil()

# Somatoria de indicadores

```python
#Soma de todas as transações da coluna 'ID Loja'
rank_loja = venda_ano['ID Loja'].value_counts()
display(rank_loja)
```

```python
#Agrupamento de produtos e soma das demais colunas
grupo = venda_ano[['Produto', 'Valor Final']].groupby('Produto').sum() #Para fazer a media dos valores de cada produto apenas substituir o .sum() por .mean()
display(grupo)
```

# Mesclar dataframe

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')#importando df
display(gerentes_df.head(3))
display(venda_ano.head(3))
```
